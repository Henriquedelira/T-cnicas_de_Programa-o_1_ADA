Unificando os DF com merge

```python
#Antes de realizar o merge, vamos avaliar se existem duplicidades nas bases:
duplicidades_df_customers = df_customers.duplicated().sum()
duplicidades_df_customers
```

```python
#Antes de realizar o merge, vamos avaliar se existem duplicidades nas bases:
duplicidades_df_geolocation = df_geolocation.duplicated().sum()
duplicidades_df_geolocation
```

```python
#Retirando as duplicidades do df_geolocation com base na primeira coluna e deixando apenas o primeiro registro encontrado
df_geolocation = df_geolocation.drop_duplicates(subset=['geolocation_zip_code_prefix'], keep='first')
```

```python
#Criando um novo df chamado df_customers_geolocation atráves de um merge a esquerda
df_customers_geolocation = pd.merge(df_customers, df_geolocation, how='left', left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix')
```

```python
df_customers_geolocation
```

```python
#Retirando colunas que possuem informações iguais
df_customers_geolocation = df_customers_geolocation.drop(columns=['geolocation_zip_code_prefix', 'geolocation_city', 'geolocation_state'])
```

```python
df_customers_geolocation
```

```python
#Criando um novo df chamado df_sellers_geolocation atráves de um merge a esquerda
df_sellers_geolocation = pd.merge(df_sellers, df_geolocation, how='left', left_on='seller_zip_code_prefix', right_on='geolocation_zip_code_prefix')
```

```python
df_sellers_geolocation
```

```python
#Retirando colunas que possuem informações iguais
df_sellers_geolocation = df_sellers_geolocation.drop(columns=['geolocation_zip_code_prefix', 'geolocation_city', 'geolocation_state'])
```

```python
df_sellers_geolocation
```

```python
#Criando um novo df chamado df_order_items_sellers atráves de um merge a esquerda
df_order_items_sellers = pd.merge(df_order_items, df_sellers_geolocation, how='left', left_on='seller_id', right_on='seller_id')
```

```python
df_order_items_sellers.info()
```

```python
#Criando um novo df chamado df_products_pt_en atráves de um merge a esquerda
df_products_pt_en = pd.merge(df_products, df_product_category, how='left', left_on='product_category_name', right_on='product_category_name')
```

```python
df_products_pt_en
```

```python
#Criando um novo df chamado df_order_items_sellers_products atráves de um merge a esquerda
df_order_items_sellers_products = pd.merge(df_order_items_sellers, df_products_pt_en, how='left', left_on='product_id', right_on='product_id')
```

```python
df_order_items_sellers_products.info()
```

```python
#Criando um novo df chamado df_orders_items atráves de um merge a esquerda
df_orders_items = pd.merge(df_order_items_sellers_products, df_orders, how='left', left_on='order_id', right_on='order_id')
```

```python
df_orders_items.info()
```

```python
df_orders_items
```

```python
#Retirando as duplicidades do df_order_reviews com base na coluna order_id e deixando apenas o primeiro registro encontrado
df_order_reviews = df_order_reviews.drop_duplicates(subset=['order_id'], keep='first')
```

```python
df_order_reviews.info()
```

```python
#Criando um novo df chamado df_orders_items atráves de um merge a esquerda
df_orders_items_reviews = pd.merge(df_orders_items, df_order_reviews, how='left', left_on='order_id', right_on='order_id')
```

```python
df_orders_items_reviews.info()
```

```python
#Devido ao fato do cliente poder realizar o pagamento em mais de uma forma em um único pedido, faremos um ajuste no df_order_payments
#Primeiro vamos realizar um groupby, somando os valores por order_id e payment_type
df_order_payments_group = df_order_payments.groupby(['order_id', 'payment_type'])['payment_value'].sum().reset_index()
```

```python
df_order_payments_group
```

```python
#Segundo vamos realizar um pivot, transfornando a coluna payment_type em vários colunas conforme o metodo de pagamento
df_order_payments_pivot = df_order_payments_group.pivot_table(index='order_id', columns='payment_type', values='payment_value', aggfunc=np.sum)
```

```python
#Ajustando o index
df_order_payments_pivot = df_order_payments_pivot.reset_index().rename_axis(None, axis=1)
```

```python
df_order_payments_pivot
```

```python
#Tratando os valores NaN
df_order_payments_pivot.fillna(0, inplace=True)
```

```python
#Criando uma nova coluna totalizando o pagamento
df_order_payments_pivot['payment_total'] = df_order_payments_pivot.apply(lambda row: row['boleto'] + row['credit_card'] + row['debit_card']  + row['not_defined']  + row['voucher'], axis=1)
```

```python
#Renomeando as colunas de tipo de pagamento
df_order_payments_pivot = df_order_payments_pivot.rename(columns={'boleto': 'payment_boleto', 'credit_card': 'payment_credit_card', 'debit_card': 'payment_debit_card', 'not_defined': 'payment_not_defined', 'voucher': 'payment_voucher'})
```

```python
df_order_payments_pivot
```

```python
#Criando um novo df chamado df_orders_items atráves de um merge a esquerda
df_orders_items_reviews_pay = pd.merge(df_orders_items_reviews, df_order_payments_pivot, how='left', left_on='order_id', right_on='order_id')
```

```python
df_orders_items_reviews_pay.info()
```

```python
df_orders_items_reviews_pay
```

```python
#Reordenando as colunas para não ter duplicidade no momento do merge
df_customers_geolocation = df_customers_geolocation.rename(columns={'geolocation_lat': 'customer_geolocation_lat', 'geolocation_lng': 'customer_geolocation_lng'})
```

```python
#Criando um novo df chamado df_orders_items atráves de um merge a esquerda
df_base_completo = pd.merge(df_orders_items_reviews_pay, df_customers_geolocation, how='left', left_on='customer_id', right_on='customer_id')
```

```python
df_base_completo.info()
```

```python
df_base_completo
```

```python
#Ajustando os valores NaN na coluna review_score
df_base_completo['review_score'] = df_base_completo['review_score'].fillna(0)
```

```python
# Criar uma máscara para classificar os review_score
mask_pessimo = df_base_completo['review_score'] < 2
mask_melhorar = (df_base_completo['review_score'] > 1) & (df_base_completo['review_score'] < 5)
mask_otimo = df_base_completo['review_score'] == 5

# Substituir os valores de acordo com a máscara
df_base_completo['Métrica_review'] = np.where(mask_pessimo, 'Péssimo', np.where(mask_melhorar, 'Melhorar', 'Ótimo' ))
```

Analisando os dados

```python
#Realizando um groupby com size para saber a contagem de reviews por estado e tipo
df_review_estado = df_base_completo.groupby(['customer_state', 'Métrica_review']).size().reset_index()
```

```python
df_review_estado = df_review_estado.rename(columns={0: 'Contagem'})
```

```python
df_review_estado
```

```python
#Segundo vamos realizar um pivot, transfornando a coluna payment_type em vários colunas conforme o metodo de pagamento
df_review_estado_pivot = df_review_estado.pivot_table(index='customer_state', columns='Métrica_review', values='Contagem', aggfunc=np.sum)
```

```python
#Ajustando o index
df_review_estado_pivot = df_review_estado_pivot.reset_index().rename_axis(None, axis=1)
```

```python
df_review_estado_pivot
```

```python
#Criando uma nova coluna com a soma dos valores
df_review_estado_pivot['total_reviews'] = df_review_estado_pivot.apply(lambda row: row['Melhorar'] + row['Péssimo'] + row['Ótimo'], axis=1)
```

```python
#Percentual de cada tipo em relação ao total por Estado
df_review_estado_pivot['perc_melhorar'] = (df_review_estado_pivot['Melhorar']  / df_review_estado_pivot['total_reviews'])*100
df_review_estado_pivot['perc_pessimo'] = (df_review_estado_pivot['Péssimo']  / df_review_estado_pivot['total_reviews'] )*100
df_review_estado_pivot['perc_otimo'] = (df_review_estado_pivot['Ótimo']  / df_review_estado_pivot['total_reviews'] )*100
```

```python
df_review_estado_pivot = df_review_estado_pivot.sort_values('perc_pessimo', ascending=False)
```

```python
df_review_estado_pivot
```

Com base no df acima, conseguimos identificar que os Estados: RR, RJ e MA possuem um percentual maior de péssimo (notas 0 e 1) relacionados ao número de reviews totais.

Plotando os dados

```python
from matplotlib import pyplot as plt
from matplotlib.lines import Line2D

# Estabelecendo o tamanho do gráfico
plt.figure(figsize=(100, 60))

# Nomeando as cores por estado para a Legenda
cores = {'AC': 'red', 'AL': 'blue', 'AM': 'green', 'AP': 'orange', 'BA': 'purple', 'CE': 'brown', 'DF': 'pink', 'ES': 'cyan', 'GO': 'magenta', 'MA': 'yellow', 'MG': 'lime', 'MS': 'teal', 'MT': 'lavender', 'PA': 'maroon', 'PB': 'olive', 'PE': 'gold', 'PI': 'navy', 'PR': 'orchid', 'RJ': 'indigo', 'RN': 'peru', 'RO': 'salmon', 'RR': 'silver', 'RS': 'tan', 'SC': 'tomato', 'SE': 'violet', 'SP': 'aqua', 'TO': 'coral'}

# Plotando o gráfico do tipo dispersão com base na coluna perc_pessimo e total_Review
scatter_plot = df_review_estado_pivot.plot(kind='scatter', x='total_reviews', y='perc_pessimo', s=32, alpha=.8, c=df_review_estado_pivot['customer_state'].map(cores))

# Criando as bordas
plt.gca().spines[['top', 'right']].set_visible(True)

# Criando a legenda
legend_elements = [Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=label) for label, color in cores.items()]

# Exibindo o gráfico e legenda
plt.legend(handles=legend_elements, title='Estados', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.show()
```

```python
#Realizando um groupby para saber qual vendedor tem o maior valor de venda
df_top_vendedor = df_base_completo.groupby(['seller_id','seller_state'])['payment_total'].sum().reset_index()
```

```python
#Filtrando os 5 primeiros vendedores por Estado
top5_vendedores_estado = df_top_vendedor[df_top_vendedor['ranking']<=5].reset_index()
```

```python
#Classificando por Estado e posição do ranking
top5_vendedores_estado = top5_vendedores_estado.sort_values(by=['seller_state', 'ranking']).reset_index()
```
